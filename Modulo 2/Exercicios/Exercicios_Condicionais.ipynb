Exercício 4 - Calculadora de Desconto. 💵 🤑

Uma loja dá 10% de desconto para compras acima de R$ 100,00. Faça um programa que:

1º - Pergunta o valor da compra.

2º - Calcula o valor final com desconto (se aplicável).

3º - Mostra o valor a pagar.

```python
### Seu código aqui

valor_compra = float(input("Digite o valor total da compra:"))

if valor_compra >100:
    desconto = valor_compra * 0.10
    valor_total = valor_compra - desconto
    print(f'Você comprou {valor_compra:.2f}, teve o desconto de {desconto}, O valor total da compra é de R$:{valor_total:.2f}')

else:
    print(f"O valor total da compra é de R$:{valor_compra:.2f}, comprou abaixo de 100 reais, não tem desconto")
```

- 5 até 6.9: "Regular"

Exercício 8 - Dia da Semana
Peça um número de 1 a 7 e mostre o dia da semana correspondente: 📅

- 1: Domingo

- 2: Segunda

- ... até 7: Sábado

```python
numero = int(input("Digite um número de 1 até 7"))

dia_correspondente = {1: 'Domingo', 2: 'Segunda', 3: 'Terça', 4: 'Quarta', 5: 'Quinta', 6: 'Sexta', 7: 'Sábado'}

if numero in dia_correspondente:

    print(f"O dia da semana correspondente a esse número é: {dia_correspondente[numero]}")

else:
    print(f"O número {numero} não corresponde a nenhum dia da semana")
```

Exercício 9 - Calculadora Simples 📏📐

- Faça uma calculadora que:

- Pede dois números

- Pede a operação (+, -, *, /)

- Dependendo da operação mostre o resultado.



```python
lista_numbers = []

operacao = input("Digite a operação matemática (+,-,*,/), que você deseja fazer:")

number1 = float(input("Digite o número que deseja calcular..."))
number2 = float(input("Digite outro número que deseja calcular..."))

lista_numbers.append(number1)
lista_numbers.append(number2)

if operacao == '+':
    soma = sum(lista_numbers)
    print(f"O resultado da operacão é: {soma}")
elif operacao == '-':
    subtracao = number1 - number2
    print(f"O resultado da operacão é: {subtracao}")
elif operacao == '*':
    multiplicacao = number1*number2
    print(f"O resultado da operacão é: {multiplicacao}")
elif operacao == '/':
    divisao = number1/number2
    print(f"O resultado da operacão é: {divisao}")
else:
    print(f"A operação {operacao}, é inválida...")
```

Exercício 10 - Sistema de Multas 🚙🚘🧾


- Velocidade máxima: 80 km/h 🚫

- Até 80: "Sem multa"

- 81-100: "Multa leve"

- Acima de 100: "Multa grave"

```python
velocidade = int(input("Digite a velocidade em que o carro passou no nosso radar:"))

if velocidade <= 80:
    print(f"A Velocidade de {velocidade}Km/h está dentro do limite da via")

elif velocidade <= 100:
    print(f"A Velocidade de {velocidade}Km/h está fora do limite da via, deve ser aplicado uma multa leve")

elif velocidade >100:
    print(f"A Velocidade de {velocidade}Km/h está fora do limite da via em grande excedência, deve ser aplicado uma Multa Grave")
```
